### Themes, Voters and Ratings will be reemoved because there are too many null values therefore it can't be used.

```python
df = df.drop(columns = ['themes','voters','rating'])
df
```

### Removing inconsitensies from the data

```python
# cleaning synopsis
df['synopsis'] = df['synopsis'].apply(synopsis_cleaning)
```

```python
df['synopsis'].sample(5)
```

```python
# cleaning title
df['title'] = df['title'].apply(title_cleaning)
```

```python
df['title'].sample(20)
```

```python
# removing punctuations from synopsis
df['synopsis'] = df['synopsis'].apply(remove_punctuation)
```

### Removing redundant statement i.e. "written by mal rewrite" from every synopsis

```python
def remove_redundant_string(text):
    return text.replace("written by mal rewrite", "").strip()

df['synopsis'] = df['synopsis'].apply(remove_redundant_string)
```

## Removeing mature Content from the genres

```python
df['genres'] = df['genres'].apply(remove_mature_content)
```

## Tagging 1 in episodes column for records which are movies

```python
condition = df['title'].str.contains('movie', case = False ) & df['episodes'].isna()

df.loc[condition, 'episodes'] = 1

df.loc[df['title'].str.contains('movie', case=False)]
```
